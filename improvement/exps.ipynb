```python
import os
import sys
import openai
import dotenv

sys.path.append("../")

dotenv.load_dotenv("../.env")
```

```python
from src.conspectum.process import process
```

```python
import httpx

ai = openai.AsyncOpenAI(
    api_key="unused",
    base_url=os.environ["AI_BASE_URL"],
    default_headers={"Authorization": f"OAuth {os.environ['AI_API_KEY']}"},
    http_client=httpx.AsyncClient(verify=False),
)
with open("/Users/raimova-al/Documents/study/master/MVP-DS/Conspectum/example/output_cut_big.wav", "rb") as audio_file:
    # language="ru" — конспект на русском, language="en" — на английском
    # language=None (по умолчанию) — автоопределение языка аудио
    tex, pdf = await process(audio_file.read(), ai, language="en")

with open("result-imp.tex", "w") as tex_file:
    tex_file.write(tex)

if pdf is not None:
    with open("result-imp.pdf", "wb") as pdf_file:
        pdf_file.write(pdf)
```
